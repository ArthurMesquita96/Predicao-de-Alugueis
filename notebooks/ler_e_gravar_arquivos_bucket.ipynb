```python
def get_all_dates(bucket_name, imobiliaria=None):

    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)

    names = pd.DataFrame({'names':[i.name for i in bucket.list_blobs()]})

    if imobiliaria:
        names = names.loc[names['names'].str.contains(imobiliaria)]
        dates_list = names['names'].apply(lambda x: x.split(' - ')[0]).tolist()

    return dates_list
```

```python
BUCKET_NAME = 'busca-apartamentos-bucket'
TEMP_FILE = 'local.csv'

dates = get_all_dates(BUCKET_NAME, 'cilar')

storage_client = storage.Client()
bucket = storage_client.get_bucket(BUCKET_NAME)

for date in dates:

    FILE_NAME = f'{date} - apartamentos - cilar.csv'
    blop = bucket.blob(blob_name= FILE_NAME).download_as_string()
    with open(TEMP_FILE, 'wb') as f:
        f.write(blop)

    df = pd.read_csv('local.csv')
    df.head()

    df = df.rename(columns={'catacteristicas_imovel':'caracteristicas_imovel'})
    df.head()

    RAW_PATH = '/tmp/{}'.format(TEMP_FILE)

    storage_client = storage.Client()
    bucket = storage_client.get_bucket(BUCKET_NAME)

    df.to_csv(RAW_PATH, index=False)
    blob = bucket.blob(FILE_NAME)
    blob.upload_from_filename(RAW_PATH)
```
