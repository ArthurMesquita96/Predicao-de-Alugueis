```python
import pandas as pd
from geopy.geocoders import Nominatim
import my_functions as utils
import time
import ast
```

```python
def try_get_latlong(endereco):
    try:
        lat_long = utils.obter_lat_long(endereco)
        print(f'endereco: {endereco}\nlat_long: {lat_long}')
        return lat_long
    except:
        pass

def get_lat_long(lat_long, coordenada):
    if lat_long:
        try: 
            lat_long_list = ast.literal_eval(lat_long)
            match coordenada:
                case 'latitude':
                    return lat_long_list[0]
                case 'longitude':
                    return lat_long_list[1]
        except:
            print('não deu certo')
            pass
```

```python
df = pd.read_pickle('params/data/df_feature_engineering.pkl')
df_lat_long = pd.DataFrame(utils.load_picked_file(name='lat_longs')).T

print(f'Data full: {df.shape[0]}')
print(f'Data Lat Long: {df_lat_long.shape[0]}')
```

```python
df_com_lat_long = pd.merge(df, df_lat_long, on = 'link', how = 'left')
```

```python
df_com_lat_long.loc[df_com_lat_long['latitude'].isna()].shape[0]
```

```python
df_com_lat_long.loc[df_com_lat_long['latitude'].isna(), 'lat_long'] = df_com_lat_long.loc[df_com_lat_long['latitude'].isna()].apply(lambda x: try_get_latlong(x['endereco']), axis=1)
```

```python
df_com_lat_long.loc[~df_com_lat_long['lat_long'].isna(), 'latitude'] = df_com_lat_long.loc[~df_com_lat_long['lat_long'].isna(), 'lat_long'].apply(lambda x: get_lat_long(str(x), 'latitude'))
df_com_lat_long.loc[~df_com_lat_long['lat_long'].isna(), 'longitude'] = df_com_lat_long.loc[~df_com_lat_long['lat_long'].isna(), 'lat_long'].apply(lambda x: get_lat_long(str(x), 'longitude'))
```

```python
df_com_lat_long[['link','latitude','longitude']].isna().sum()
```

```python
dict_lat_long = df_com_lat_long[['link','latitude','longitude']].to_dict('index')
utils.save_picked_file(file=dict_lat_long, name='lat_longs')
```
